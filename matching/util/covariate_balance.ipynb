```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# sns.set(rc={'figure.figsize':(16,10)}, font_scale=1.3)
```

```python
coraal_full = pd.read_csv("CORAAL_transcripts_clean.csv")
nsp_full = pd.read_csv("nsp_transcript_clean.csv")
coraal_matched = pd.read_csv("coraal_matched.csv")
nsp_matched = pd.read_csv("nsp_matched.csv")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

for x, variable in enumerate(["age", "duration"]):
    sns.kdeplot(coraal_full[variable], label='CORAAL', ax=ax[0][x])
    sns.kdeplot(nsp_full[variable], label='NSP', ax=ax[0][x])
    ax[0][x].legend()

for x, variable in enumerate(["age", "duration"]):
    sns.kdeplot(coraal_matched[variable], label='CORAAL', ax=ax[1][x])
    sns.kdeplot(nsp_matched[variable], label='NSP', ax=ax[1][x])
    ax[1][x].legend()

fig.text(0.05, 0.6, 'Pre-matching', rotation='vertical')
fig.text(0.05, 0.2, 'Post-matching', rotation='vertical')
```

```python
fig, ax = plt.subplots(2, 1, figsize=(3, 8))
fig.subplots_adjust(hspace=0.3)

# pre-matching
coraal_full_gender = coraal_full['gender'].value_counts(normalize=True) * 100
nsp_full_gender = nsp_full['gender'].value_counts(normalize=True) * 100
bar_width = 0.35
bar_pos_coraal = range(len(coraal_full_gender))
bar_pos_nsp = [pos + bar_width for pos in bar_pos_coraal]
ax[0].bar(bar_pos_coraal, coraal_full_gender, width=bar_width, label='CORAAL')
ax[0].bar(bar_pos_nsp, nsp_full_gender, width=bar_width, label='NSP')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Pre-Matching Percentage')
ax[0].set_xticks([pos + bar_width / 2 for pos in bar_pos_coraal])
ax[0].set_xticklabels(coraal_full_gender.index)
ax[0].legend()

# post-matching
coraal_matched_gender = coraal_matched['gender'].value_counts(normalize=True) * 100
nsp_matched_gender = nsp_matched['gender'].value_counts(normalize=True) * 100
bar_width = 0.35
bar_pos_coraal = range(len(coraal_matched_gender))
bar_pos_nsp = [pos + bar_width for pos in bar_pos_coraal]
ax[1].bar(bar_pos_coraal, coraal_matched_gender, width=bar_width, label='CORAAL')
ax[1].bar(bar_pos_nsp, nsp_matched_gender, width=bar_width, label='NSP')
ax[1].set_xlabel('Gender')
ax[1].set_ylabel('Pre-Matching Percentage')
ax[1].set_xticks([pos + bar_width / 2 for pos in bar_pos_coraal])
ax[1].set_xticklabels(coraal_matched_gender.index)
ax[1].legend(loc='lower right')
```
